# Diamonds Dataset Analysis
Exploring the relationship between carat (size) and price for diamonds with clarity 'SI2' and color 'E'

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load and filter data
df = pd.read_csv(r'Dataset\diamonds.csv')
diamonds = df[(df['clarity'] == 'SI2') & (df['color'] == 'E')].copy()

print(f"Total diamonds: {len(df)}")
print(f"Filtered diamonds (SI2 clarity, E color): {len(diamonds)}")
print(f"Carat range: {diamonds['carat'].min():.2f} - {diamonds['carat'].max():.2f}")
print(f"Price range: ${diamonds['price'].min():.0f} - ${diamonds['price'].max():.0f}")
```

```python
# Create scatter plot
fig, ax = plt.subplots(figsize=(14, 8))

# Color map based on carat values
scatter = ax.scatter(diamonds['carat'], diamonds['price'], 
                     c=diamonds['carat'], 
                     cmap='viridis', 
                     s=100, 
                     alpha=0.7,
                     edgecolors='black',
                     linewidth=0.5)

# Customize chart
ax.set_title('Diamond Price vs Carat Weight\n(Clarity: SI2, Color: E)', 
             fontsize=18, fontweight='bold', pad=20, color='#1a1a1a')
ax.set_xlabel('Carat Weight (Size)', fontsize=14, fontweight='bold')
ax.set_ylabel('Price ($)', fontsize=14, fontweight='bold')

# Format axes
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${int(x):,}'))
ax.grid(True, linestyle='--', alpha=0.3, color='gray')
ax.set_axisbelow(True)

# Customize ticks
ax.tick_params(axis='both', labelsize=11)

# Add colorbar with carat label
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Carat Weight', fontsize=12, fontweight='bold')
cbar.ax.tick_params(labelsize=11)

plt.tight_layout()
plt.show()

print("Scatter plot created successfully!")
```

```python
# Summary statistics
print('\n=== Summary Statistics (SI2 Clarity, E Color) ===')
print(f"\nNumber of diamonds: {len(diamonds)}")
print(f"\nCarat Weight:")
print(f"  Mean: {diamonds['carat'].mean():.3f}")
print(f"  Median: {diamonds['carat'].median():.3f}")
print(f"  Std Dev: {diamonds['carat'].std():.3f}")
print(f"\nPrice:")
print(f"  Mean: ${diamonds['price'].mean():.2f}")
print(f"  Median: ${diamonds['price'].median():.2f}")
print(f"\nCorrelation between Carat and Price: {diamonds['carat'].corr(diamonds['price']):.4f}")
```
