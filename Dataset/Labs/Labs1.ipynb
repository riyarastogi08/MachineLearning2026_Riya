# European Population Analysis (1952-2007)
Comparing populations of France, UK, Italy, Germany, and Spain

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and prepare data
df = pd.read_csv(r'C:\Users\Riya Rastogi\OneDrive\Desktop\MachineLearning2026_Riya\MachineLearning2026_Riya\Dataset\gapminderDataFiveYear.csv')
countries = ['France', 'United Kingdom', 'Italy', 'Germany', 'Spain']
data = df[df['country'].isin(countries)].pivot(index='year', columns='country', values='pop')
```

```python
# Create and customize line chart
fig, ax = plt.subplots(figsize=(14, 8))

colors = {'France': '#1f77b4', 'United Kingdom': '#ff7f0e', 'Italy': '#2ca02c', 'Germany': '#d62728', 'Spain': '#9467bd'}

for country in countries:
    ax.plot(data.index, data[country], marker='o', linewidth=2.5, markersize=6, label=country, color=colors[country], alpha=0.8)

# Customize chart
ax.set_title('European Population Trends (1952-2007)', fontsize=18, fontweight='bold', pad=20, color='#2c3e50')
ax.set_xlabel('Year', fontsize=14, fontweight='bold')
ax.set_ylabel('Population in lakhs', fontsize=14, fontweight='bold')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x/1e6)}L'))
ax.grid(True, linestyle='--', alpha=0.3)
ax.set_xticks(range(1952, 2008, 5))
ax.tick_params(axis='x', labelsize=11, rotation=45)
ax.tick_params(axis='y', labelsize=11)
ax.legend(fontsize=12, loc='upper left', framealpha=0.95, edgecolor='#2c3e50', fancybox=True, shadow=True)

plt.tight_layout()
plt.show()
```
