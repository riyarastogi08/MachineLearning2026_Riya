# Linear Regression Example with Line of Best Fit 


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

X_b = np.c_[np.ones((100, 1)), X]

theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

print(f"Intercept (Bias): {theta_best[0][0]:.2f}")
print(f"Coefficient (Weight): {theta_best[1][0]:.2f}")
```

```python
X_new = np.array([[0], [2]])

X_new_b = np.c_[np.ones((2, 1)), X_new]

y_predict = X_new_b.dot(theta_best)

plt.figure(figsize=(10, 6))
plt.plot(X, y, "b.", label="Data Points") 
plt.plot(X_new, y_predict, "r-", linewidth=2, label="Line of Best Fit") 

plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression - Line of Best Fit")
plt.legend()
plt.axis([0, 2, 0, 15])
plt.grid(True)
plt.show()
```
